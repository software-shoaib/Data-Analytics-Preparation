```python
data1 = pd.read_csv("weight-height.csv")
```

```python
data = data1[['Height', 'Weight']]
```

```python
data
```

```python
height = data.Height
weight = data.Weight
```

```python
plt.figure(figsize=(40,40))
plt.scatter(height,weight)
```

```python
X = data[['Height']]
y = data[['Weight']]
```

```python

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create and fit the linear regression model
height_weight_regressor = LinearRegression()
height_weight_regressor.fit(X_train, y_train)
y_pred_train = height_weight_regressor.predict(X_train)

# Flatten the y_train and y_pred_train arrays
y_train_flat = y_train['Weight'].values.flatten()
y_pred_train_flat = y_pred_train.flatten()

# Create the DataFrame with flattened arrays
df = pd.DataFrame({'y_train': y_train_flat, 'y_pred_train': y_pred_train_flat})

# Now df contains the flattened arrays in a DataFrame
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train_flat,y_pred_train_flat)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train_flat,y_pred_train_flat)
```

```python
height_weight_regressor.coef_
```

```python
height_weight_regressor.intercept_
```

```python
user_height = float(input("enter your height, I will tell your age: "))
print(f"your according to your height {user_height} your weight is {height_weight_regressor.predict([[user_height]])}")
```
