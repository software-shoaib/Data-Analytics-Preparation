```python
# Merge by default merges data on overlapping columns. 
#If more than one overlapping columns present than it selects the first one by default.
# If i will supply overlapping columns name ourselves then i will not select the column by default.
# since any of the sales man name of  df1 has only one repeated member  
# Thats why only one reocrd is merged and shown in the frame.
df1.merge(df2, on='sales_man_name')
```

```python
df1  = pd.read_excel('test_data.xlsx',sheet_name=0)
df2  = pd.read_excel('test_data.xlsx',sheet_name=1)
df3  = pd.read_excel('test_data.xlsx',sheet_name=2)
```

```python
df3.sales_man_name = df3.sales_man_name.apply(lambda x: x.strip())
df3.Region = df3.Region.apply(lambda x: x.strip())
```

```python
df3.rename(columns={'Region ':'Region', 'Amount ':'Amount'}, inplace=True)
df3.columns
```

```python
df3.columns = ['sales_man_name', 'Region', 'Sales', 'Amount'
```

```python
df3.Region = df3.Region.apply(lambda x: x.strip())
```

```python
df3
```

```python
df3.rename(columns={'sales_man_name':'sales_man'},inplace=True)
```

```python
df3.merge(df1,left_on='sales_man', right_on='sales_man_name').drop('sales_man_name',axis=1)
```

```python
df3.merge(df1,how='left', left_on='sales_man', right_on='sales_man_name').drop('sales_man_name',axis=1)
```

```python
df3.merge(df1,how='right', left_on='sales_man', right_on='sales_man_name').drop('sales_man_name',axis=1)
```
